```python
from flask import Flask, request, jsonify
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
import nltk
import gspread
from flask_cors import CORS


# Initialize Flask
app = Flask(__name__)
CORS(app)

# Initialize NLTK
nltk.download('wordnet')

# Define aspect terms and negative keywords
aspects = ["doctor care", "nurse care", "billing process", "wait time", "hospitality"]
negative_keywords = ["below expectations", "average", "unimpressed", "not", "substandard", "unsatisfactory", "poor", "excessive", "mixed experience", "disappointed", "frustrating", "subpar", "very poor", "lacking", "much to be desired", "bad", "very bad", "needs improvement", "tiresome", "below par", "could be improved", "negative"]

# Define subaspects
subaspects = {
    "hospitality": ["diet", "communication", "hospital facilities", "ward", "hospital care"],
    # Add other aspects and their subaspects as needed
}

# Initialize the SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Preprocess function to tokenize and lemmatize words
def preprocess(text):
    tokens = word_tokenize(text)
    lemmatizer = nltk.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return " ".join(lemmatized_tokens)

# Route for sentiment analysis
@app.route('/analyze-sentiment', methods=['POST'])
def analyze_sentiment():
    try:
        # Get the recorded text data from the request
        data = request.get_json()
        recorded_text = data['recorded_text']

        # Preprocess the recorded text
        preprocessed_text = preprocess(recorded_text)

        # Initialize a list to store the predicted sentiments
        predicted_sentiments = []

        # Perform aspect-based sentiment analysis on the provided text
        text = preprocessed_text

        for aspect in aspects:
            aspect_lower = aspect.lower()
            if any(keyword in text for keyword in negative_keywords) and aspect_lower in text:
                sentiment = "negative"
            else:
                sentiment = "positive"

            # Check for subaspects
            if aspect_lower in subaspects:
                subaspect_list = subaspects[aspect_lower]
                for subaspect in subaspect_list:
                    if subaspect.lower() in text:
                        sentiment = "negative"
                        break  # Break the loop if a subaspect is found

            predicted_sentiments.append(sentiment)

        # Convert the list of sentiments to a string separated by colons
        sentiments_string = ":".join(predicted_sentiments)

        # Split the sentiments string by colons to get individual sentiments
        individual_sentiments = sentiments_string.split(":")

        # Authenticate using the credentials file
        gc = gspread.service_account(filename="C:/Users/u1147359/Downloads/h2023-398212-5160d189a8cb.json")

        # Open the Google Sheet by title or URL
        sheet_url = "https://docs.google.com/spreadsheets/d/13sJWZXQY2CqneHTUtbxzwhNFQ_vnMZtKlRi2xZvXJFs/edit?usp=sharing"
        spreadsheet = gc.open_by_url(sheet_url)

        # Select the desired sheet (by title or index)
        worksheet = spreadsheet.sheet1

        # Create a list to store the data to insert into the sheet
        data_to_insert = [recorded_text] + individual_sentiments

        # Add the data to the sheet as a new row
        worksheet.append_row(data_to_insert)

        return jsonify({'message': 'Data inserted into Google Sheets.'})
    except Exception as e:
        return jsonify({'error': str(e)})

if __name__ == '__main__':
    app.run(debug=False)
```

```python
from flask import Flask, request, jsonify
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
import nltk
import gspread
from flask_cors import CORS
import spacy

# Initialize Flask
app = Flask(__name__)
CORS(app)

# Initialize NLTK
nltk.download('wordnet')

# Initialize spaCy model
import spacy


# Load the model
nlp = spacy.load("en_core_web_sm")





# Define aspect terms and negative keywords
aspects = ["Doctor Care", "Nurse Care",  "Support Staff Care","Billing Process and Financial Transparency","Communication", "Wait Time", "Hospital Amenities","Patient Hospitality"]



# Define aspect terms and their aliases
aspect_mapping = {
    "Doctor Care": ["doctor", "physician", "surgeon", "doctors", "physicians", "surgeons", "specialist","surgical team"],
    "Nurse Care": ["nurse", "nurses", "sister", "medical staff"],
    "Support Staff Care": ["physiotherapists", "technicians", "therapists", "support staff", "staff",
                           "physiotherapist", "technician", "therapist", "rehabilitation", "ward boy","department","teams","management team","personnel","unit"],
    "Billing Process and Financial Transparency": ["billing process", "medical bill clarity","billing","bill"],
    "Communication": ["procedure explaination", "treatment plan information", "medication instruction","instruction","instructions"
                      "communication", "information","education","direction","sign","process","planning","scheduling","schedule","follow up","follow-up"],
    "Wait Time": ["waiting", "wait","appointment","response time","explained"],
    "Hospital Amenities": ["pharmacy", "equipment", "parking", "facilities", "infection control", "waiting area",
                           "waiting room","cafeteria","appointment system","internet", "wi-fi","wifi","net","shop","interpreter","interpreters"
                           "center","transport","chaplain","breastfeeding","patient portal","toys","shop's","playroom"],
    "Patient Hospitality": ["infection control", "compassion", "diet", "cleanliness", "empathy", "care", "safe",
                            "safety", "ward", "services", "dietary", "bed","caring","attentive","security","unit","visiting","direction",
                            "sign","process","service","follow up","follow-up","unfriendly","stay","cafeteria","visitation","program","hygiene","cafeteria's"]
}

# Negative keywords
negative_keywords = ["unappetizing", "limited", "ineffective", "unhelpful", "little information", "nightmare",
                     "challenging", "confusing", "unwelcoming", "uncomfortable", "slow response", "slow to respond",
                     "chaotic", "outdated", "poorly", "inattentive", "unsanitary", "disintrested", "dissmissive",
                     "lacked", "didn't", "for hours", "rude", "inattentive", "below expectations", "average",
                     "unimpressed", "not ", "substandard", "unsatisfactory", "poor", "mixed experience", "disappointed",
                     "frustrating", "subpar", "very poor", "lacking", "much to be desired", "bad", "very bad",
                     "needs improvement", "tiresome", "below par", "could be improved", "negative","too long","non","rude",
                     "inadequate","errors","error","impatient","long wait","killed","death","died","invaluable","expensive",
                     "inconvenient","disorganized","outdated","stressful","stress","unapproachable","incorrect","hard",
                     "unsupportive","unreliable","pricy","disorganised","unacceptable","too strict","disrepair","unprofessional",
                     "unclear","unpredictable","incomplete","unreasonable","glitch","glitches","underutilized","often cold","excessive",
                     "depleted","inflexible","unresponsive","prohibitive","unpalatable","side effects","cramped","overcooked",
                     "questionable","unaffordable","cluttered","time was slow","rescheduled","disorganized","diffult","too salty",
                     "bland","slow","dismissive","overlooked","frequently down","need of","greasy","rushed","unsafe","tastless",
                     "rescheduled","hindering","overpriced","hardship","dangerous","uninviting","dissatisfaction","intolerable",
                     "distress","antiquated","burden"]




# Preprocess function to tokenize and lemmatize words
def preprocess(text):
    tokens = word_tokenize(text)
    lemmatizer = nltk.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return " ".join(lemmatized_tokens)

# Route for sentiment analysis
@app.route('/analyze-sentiment', methods=['POST'])
def analyze_sentiment():
    try:
        # Get the recorded text data from the request
        data = request.get_json()
        recorded_text = data['recorded_text']

        # Preprocess the recorded text
        preprocessed_text = preprocess(recorded_text)

        # Initialize a list to store the predicted sentiments
        predicted_sentiments = []

        # Perform aspect-based sentiment analysis on the provided text
        doc = nlp(preprocessed_text.lower())  # Convert text to lowercase for case-insensitivity

        for sent in doc.sents:
            for aspect, aliases in aspect_mapping.items():
                for alias in aliases:
                    if alias in sent.text:
                        for keyword in negative_keywords:
                            if keyword in sent.text:
                                predicted_sentiments.append((aspect, "Negative"))
                                break

        if not predicted_sentiments:
            predicted_sentiments.append(("Overall Sentiment", "Positive"))

        # Convert the list of sentiments to a dictionary
        sentiments_dict = dict(predicted_sentiments)

        # Authenticate using the credentials file
        gc = gspread.service_account(filename="C:/Users/u1147359/Downloads/h2023-398212-5160d189a8cb.json")

        # Open the Google Sheet by title or URL
        sheet_url = "https://docs.google.com/spreadsheets/d/13sJWZXQY2CqneHTUtbxzwhNFQ_vnMZtKlRi2xZvXJFs/edit?usp=sharing"
        spreadsheet = gc.open_by_url(sheet_url)

        # Select the desired sheet (by title or index)
        worksheet = spreadsheet.sheet1

        # Create a list to store the data to insert into the sheet
        data_to_insert = [recorded_text] + [sentiments_dict.get(aspect, "Positive") for aspect in aspects]

        # Add the data to the sheet as a new row
        worksheet.append_row(data_to_insert)

        return jsonify({'message': 'Data inserted into Google Sheets.'})
    except Exception as e:
        return jsonify({'error': str(e)})

if __name__ == '__main__':
    app.run(debug=False,port=5001)
```
